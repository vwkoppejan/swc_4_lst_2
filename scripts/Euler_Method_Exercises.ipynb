```python
#Euler method
import matplotlib.pyplot as plt
import numpy as np
DT = [ 0.001, 0.01, 0.02]
p1 = ['--b', '.--b', 'v--b']
p2 = ['--r', '.--r', 'v--r']
i = 0
for dt in DT:
    t, u, v = 0, 1, 0
    T = [t]; U = [u]; V = [v]
    while t<0.25:
        dudt = 98*u+198*v
        dvdt = -99*u-199*v
        u = u + dudt*dt
        v = v + dvdt*dt
        t += dt

        T.append(t)
        U.append(u)
        V.append(v)
    plt.plot(T, U, p1[i])
    plt.plot(T, V, p2[i])
    i += 1

```

```python
#Backward Euler method
import matplotlib.pyplot as plt
import numpy as np
dt = 0.2
t = np.linspace(0,1, 1000)   
f = np.exp(-10*t)
plt.plot(t,f,'k', label ='Exact')

t, y, e = 0, 1, 0
T = [t]; Y = [y]; E_rel = [e_rel]; E_abs = [e_abs]
while t<1:
    y = y/(1+10*dt)
    exact = np.exp(-10*t)
    e_abs =abs(y - exact)
    e_rel =100*e_abs/exact
    t += dt
    T.append(t)
    Y.append(y)
    E_rel.append(e_rel)
    E_abs.append(e_abs)
        

        
plt.figure(1)
plt.plot(T, Y, 'r', label = 'Approximation')
plt.legend(loc ="upper right")
plt.ylabel('y')
plt.xlabel('t')
plt.figure(2)
plt.plot(T, E_rel)
plt.title('Relative error')
plt.ylabel('%')
plt.xlabel('t')
plt.figure(3)
plt.plot(T, E_abs)
plt.title('Absolute error')
plt.ylabel('Error')
plt.xlabel('t')

plt.show()
```

```python
#Forward Euler method
import matplotlib.pyplot as plt
import numpy as np
dt = 0.2
t = np.linspace(0,1, 1000)   
f = np.exp(-10*t)
plt.plot(t,f,'k', label ='Exact')

t, y, e = 0, 1, 0
T = [t]; Y = [y]; E_rel = [e_rel]; E_abs = [e_abs]
while t<1:
    y = y*(1-10*dt)
    exact = np.exp(-10*t)
    e_abs =abs(y - exact)
    e_rel =100*e_abs/exact
    t += dt
    T.append(t)
    Y.append(y)
    E_rel.append(e_rel)
    E_abs.append(e_abs)
        

        
plt.figure(1)
plt.plot(T, Y, 'r', label = 'Approximation')
plt.legend(loc ="upper right")
plt.ylabel('y')
plt.xlabel('t')
plt.figure(2)
plt.plot(T, E_rel)
plt.title('Relative error')
plt.ylabel('%')
plt.xlabel('t')
plt.figure(3)
plt.plot(T, E_abs)
plt.title('Absolute error')
plt.ylabel('Error')
plt.xlabel('t')

plt.show()
```
